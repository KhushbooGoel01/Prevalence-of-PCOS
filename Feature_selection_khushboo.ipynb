# Feature Selection For The Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df=pd.read_csv("PCOS_changed_clean2.csv")
```

```python
df.describe().T
```

```python
# Find rows containing '.' in any column
error_rows = df[df.applymap(lambda x: x == '.').any(axis=1)]

# Print the problematic rows
print(error_rows)
```

```python
plt.figure(figsize=(50,50))
sns.heatmap(df.corr(), annot=True)
plt.show()
```

```python
corr_features=df.corrwith(df["PCOS (Y/N)"],method='pearson').abs().sort_values(ascending=True)
corr_features=corr_features[corr_features>0.4].index
corr_features
```

```python
df_f=df[corr_features]
df_f.head()
```

```python
plt.figure(figsize=(6,5))
sns.heatmap(df_f.corr(), annot=True)
plt.show()
```
