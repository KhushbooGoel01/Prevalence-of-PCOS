# Importing Necessary Modules

```python
import sys
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use("ggplot")
plt.rcParams['figure.figsize'] = (12, 8)
import seaborn as sns
sns.set(style='whitegrid', color_codes=True)
import warnings
warnings.filterwarnings('ignore')
```

# Load Dataset

```python
df_f = pd.read_csv('PCOS_changed.csv')
df_f.head(12).T
```

```python
df_f.info()
```

# Explore Dataset

# Check For Null Values

```python
df_f.isnull()
```

```python
df_f.isnull().sum()
```

```python
# Drop the column containg almost all null and uninterpretable values
df_f = df_f.drop(columns='Unnamed: 44')
# Drop unnecessary colunms
df_f = df_f.drop(columns=['Sl. No','Patient File No.'])
```

```python
df_f.head(12).T
```

# Drop Rows with null values

```python
df_f.dropna(axis=0, inplace=True)
```

```python
df_f.isnull().sum()
```

```python
df_f.info()
```

# Calculate new relevant columns


```python
df_f
```

```python
# calculating BMI
df_f['BMI'] = (df_f['Weight (Kg)']/(df_f['Height(Cm) '])**2)*10000
df_f
```

```python
# getting FSH LH ratio
df_f.loc[:,"FSH/LH"] = df_f.loc[:,"FSH(mIU/mL)"] / df_f.loc[:,"LH(mIU/mL)"];
df_f.loc[:,"FSH/LH"] = df_f.loc[:,"FSH/LH"].round(2);
```

```python
# getting waist hip ratio as that is also imortant indicator
df_f.loc[:,"Waist:Hip Ratio"] = df_f.loc[:,"Waist(inch)"] / df_f.loc[:,"Hip(inch)"]
df_f.loc[:,"Waist:Hip Ratio"] = df_f.loc[:,"Waist:Hip Ratio"].round(2)
```

```python
df_f.head(12)
```

```python
df_f.info()
```

## Count

```python
print(df_f['PCOS (Y/N)'].value_counts())
plt.figure(figsize=(6, 6))
sns.countplot(
    x='PCOS (Y/N)',
    data=df_f
);
```

```python
print(df_f['Cycle length(days)'].value_counts())
plt.figure(figsize=(6, 6))
sns.countplot(
    x='Cycle length(days)',
    data=df_f
);
```
